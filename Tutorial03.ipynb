```python
import numpy as np
# 1. Create a numpy ar{ray of size 2 x 2
arr=np.array([[1,2],[3,4]])
print("2x2 Array: \n", arr)
```

```python
# 2. Create a random numpy array of size 3x4
randarr=np.random.randint(0,51,size=(3,4))
print("\n Random Array: \n",randarr)
```

```python
# 3. Create a numpy array of ones and a numpy array of zeroes
ones=np.ones((3,3))
zero=np.zeros((3,3))
print("\nArray of Ones: \n", ones)
print("\nArray of Zeros: \n",zero)
```

```python
# 4.Create a random numpy array of size 6x3 and transpose it
arr=np.random.randint(0,51,size=(6,3))
transpose_arr=arr.T
print("\nArray before transpose: \n",arr)
print("\nArray after transpose: \n",transpose_arr)
```

```python
#5. Create two random numpy arrays of size n x p and p x m and compute the dot product
n,p,m=2,3,4
A=np.random.rand(n,p)
B=np.random.rand(p,m)
dot_product=np.dot(A,B)
print("\nMatrix A (nxp):\n",A)
print("\nMatrix B (pxm):\n",B)
print("\n Dot product (nxm):\n",dot_product)
```
